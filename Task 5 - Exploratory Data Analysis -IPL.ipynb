# Team with most no. of wins

```python
teams = matches['winner'].value_counts()
teams
```

```python
plt.figure(figsize = (15,8))
sns.barplot( y = matches['winner'].value_counts().index , x = matches['winner'].value_counts().values, palette= 'tab10')
plt.title("Number of Matches win by team", fontsize= 30)
plt.style.use('dark_background')
plt.xlabel("Number of matches won" , fontsize = 15)
plt.ylabel("Teams", fontsize=15 )
plt.show()
```

Mumbai Indians Won Maximum Number of Matches(109) followed by Chennai Super Kings(100) and Kolkata Knight Riders(92).



# Winner teams in each season


```python
season_winner = matches.drop_duplicates(subset=["season"], keep="last")[["season", "winner"]].reset_index(drop=True)
season_winner
```

```python
# Number of season win by team 
plt.figure(figsize = (12,5))
sns.countplot('winner' , data = season_winner)
plt.title("Season win by teams" , fontsize = 30)
plt.xlabel("teams" , fontsize  = 15)
plt.ylabel("count" , fontsize = 15)
plt.xticks(rotation = 90)
plt.show()
```

# Matches where team won by highest run

```python
matches.iloc[matches['win_by_runs'].idxmax()]
```

```python
plt.figure(figsize=(15,8))
win_max = matches.groupby(["winner"], as_index=False)["win_by_runs"].max()
sns.barplot(y = win_max["winner"], x = win_max["win_by_runs"], palette= 'tab10')
plt.title("Teams won by highest run" , fontsize = 30 )
plt.xlabel("win by runs", fontsize = 15)
plt.ylabel("teams" , fontsize = 15)
plt.show()
```

# Matches where team won by highest wickets

# Statistics of teams

```python
x=matches.team1.value_counts() + matches.team2.value_counts()
y=matches.winner.value_counts()
z=x-y

team_stats = pd.DataFrame({"Total matches played":x, "Total matches won":y,"Total matches lost":z})
team_stats =team_stats.reset_index()
team_stats.rename(columns={"index":"Teams"}, inplace=True)
team_stats
```

# Matches won by Mumbai Indians per season

```python
plt.figure(figsize=(15,8))
x = matches[matches["winner"]=="Mumbai Indians"].groupby(["season"]).count()
x["winner"].plot(kind="line")
plt.title("matches won by Mumbai Indians per season" , fontsize = 30 )
plt.xlabel("season" , fontsize = 15)
plt.show()
```

# Conclusion: most successful team is Mumbai Indians

# Most Successful players

# Batsman scorecard

```python
scorecard = deliveries.groupby("batsman").agg({'ball':'count','batsman_runs':'sum'})
scorecard = scorecard.sort_values(["ball", "batsman_runs"], ascending=False)
scorecard.head(10)
```

# strike rate of batsman

```python
scorecard['strikes_rate'] = scorecard['batsman_runs']/scorecard['ball']*100
x  = scorecard.round(2).head(10)
x = x.sort_values(['ball','batsman'], ascending = True)
x
```

# Conclusion : Most successfull player is CH Gayle
